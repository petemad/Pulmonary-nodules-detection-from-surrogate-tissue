```python
import pylidc as pl
import numpy as np
```

```python
mal_scans = pl.query(pl.Scan).join(pl.Annotation).filter(pl.Annotation.malignancy == 5)
ben_scans = pl.query(pl.Scan).join(pl.Annotation).filter(pl.Annotation.malignancy == 1)
```

```python
print(mal_scans.count())
print(ben_scans.count())
```

```python
mal_scans[100] == mal_scans.first()
```

```python
def extract_surrounding_tissue (scan, block_size, NODULE_NUM = 0):
    volume = scan.to_volume()
    nodules = scan.cluster_annotations()
    nodule_location = nodules[NODULE_NUM][NODULE_NUM].bbox_matrix()
    x = nodule_location[0]
    y = nodule_location[1]
    z = nodule_location[2]
    #Left
    v1 = volume[x[0]-block_size:x[0], y[0]: y[0]+block_size, z[0]: z[0]+block_size]
    #Right
    v2 = volume[x[1]:x[1]+block_size, y[0]: y[0]+block_size, z[0]: z[0]+block_size]
    #Down
    v3 = volume[x[0]:x[0]+block_size, y[0]-block_size: y[0], z[0]: z[0]+block_size]
    #Up
    v4 = volume[x[0]:x[0]+block_size, y[1]: y[1]+block_size, z[0]: z[0]+block_size]
    #Out
    v5 = volume[x[0]:x[0]+block_size, y[0]: y[0]+block_size, z[0]-block_size: z[0]]
    #In
    v6 = volume[x[0]:x[0]+block_size, y[0]: y[0]+block_size, z[1]: z[1]+block_size]
    
    return [v1, v2, v3, v4, v5, v6]
```

```python
BLOCK_SIZE = 64
n_saved_volumes = 0
n_cont = 0
for i in range(mal_scans.count()):
    scan = mal_scans[i]
    try:
        six_volumes = extract_surrounding_tissue(scan, BLOCK_SIZE)
    except:
        n_cont += 1
        continue
    for j in range(len(six_volumes)):
        np.save('./M_Data_64/v' + str(n_saved_volumes), six_volumes[j], False)
        n_saved_volumes += 1
```

```python
BLOCK_SIZE = 64
n_saved_volumes = 0
n_cont = 0
for i in range(ben_scans.count()):
    scan = ben_scans[i]
    try:
        six_volumes = extract_surrounding_tissue(scan, BLOCK_SIZE)
    except:
        n_cont += 1
        continue
    for j in range(len(six_volumes)):
        np.save('./B_Data_64/v' + str(n_saved_volumes), six_volumes[j], False)
        n_saved_volumes += 1
```
